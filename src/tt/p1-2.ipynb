```kotlin
import kotlin.math.exp
import kotlin.math.pow

fun f(t: Int, lambda: Number): Double {
    return Math.E.pow(lambda.toDouble() * t)
}
```

```kotlin
val time = 0..10
val lambdas = listOf(0.1, 0.2, 0.3, 0.4)

val views = lambdas.map { lambda ->
    time.map { time ->
        round(f(time, lambda))
    }
}
```

```kotlin
import org.jetbrains.kotlinx.dataframe.api.dataFrameOf
import org.jetbrains.letsPlot.commons.values.Colors

val time = 0..10
val lambdas = listOf(0.1, 0.2, 0.3, 0.4)

val c = mapOf(
    0 to Color.GREY,
    1 to Color.GREEN,
    2 to Color.YELLOW,
    3 to Color.RED,
)

plot {
    views.forEachIndexed { i, v ->
        line {
            x(time)
            y(v)
            color = c[i]
        }
    }

    layout {
        xAxisLabel = "Time"
        yAxisLabel = "Views"
    }
}
```
